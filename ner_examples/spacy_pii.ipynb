##### Create a Python script to detect PII using Spacy.

```python
import spacy

# Load the pre-trained model
nlp = spacy.load('en_core_web_sm')

# Define a sample text containing PII
text = """John Doe lives at 1234 Elm Street, Springfield, and his phone number is (555) 123-4567.
          His email address is john.doe@example.com."""

# Process the text with the NLP model
doc = nlp(text)

# Print the detected entities and their types
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```
